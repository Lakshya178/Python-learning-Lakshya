```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

### Basic graph Line Chart

```python
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]

#Making changes in the figure size and the DPI
plt.figure(figsize=(5,3), dpi = 100)

#plt.plot(x,y, label='2x', color='#236548', linewidth=2, marker='.', markersize=10, markeredgecolor = 'black', linestyle='--')

#shorthand notation
#fmt = '[color][marker][lineStyle]'
plt.plot(x,y,'bo-', label='2x')

#Second line
x2 = np.arange(0,4.5,0.5)

#Plotting both the lines
plt.plot(x2[:6],x2[:6]**2, color='red', label = 'x2^2')

#Plotting the remaining graph as a dotted line
plt.plot(x2[5:],x2[5:]**2, color='red', label = 'x2^2 projection', linestyle='--')

#Labelling with changes in font and font size
plt.xlabel('X Axis', fontdict={'fontname':'Times New Roman'}, fontsize=20)
plt.ylabel('Y Axis', fontdict={'fontname':'Times New Roman'}, fontsize=20)
plt.title('Lakshya', fontdict={'fontname':'Times New Roman'}, fontsize=20)

#Putting a legend on the graph
plt.legend()

plt.xticks(x)
plt.yticks(y)

#Saving a graph, 'dpi' is used for saving image at higher DPI, so it will not look pixelated when zoomed in
#plt.savefig('savedGraph300.png', dpi=300)

plt.show()
```
