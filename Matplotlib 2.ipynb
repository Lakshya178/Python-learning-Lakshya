```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.figure(figsize=(10,7))

plt.title('Gas Prices Over Time in USD')
plt.xlabel('Year')
plt.ylabel('Gas price in USD')

plt.xticks(gas.Year, rotation=75)

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
# plt.plot(gas.Year, gas.Canada,'r.-' ,label='Canada')
# plt.plot(gas.Year, gas['South Korea'],'g.-', label='South Korea')
# plt.plot(gas.Year, gas.Australia,'m.-',  label='Australia')
# plt.plot(gas.Year, gas.France,'c.-', label='France')
# plt.plot(gas.Year, gas.Germany,'k.-', label='Germany')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label = country)
    
    
plt.legend(bbox_to_anchor=(1,1))

plt.show()
```
